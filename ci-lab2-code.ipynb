```python
import numpy as np
import matplotlib.pyplot as plt

# define a function
def sphere(x):
    return x[0] * x[0] + x[1] * x[1]

def sphere_derivative(x):
    return x[0] + x[1]

# Local search (LS)
def LS(func, func_derivative, S, step_size=0.1, t_max=100):
    t = 0
    # Find starting point
    x = np.random.uniform(low=S[0], high=S[1])
    y = np.random.uniform(low=S[0], high=S[1])
    print([x,y])
    # Looping
    while t < t_max:
        z = func([x,y])
        print("t =", t, ": f(", x, ") =", z)
        q = func_derivative([x,y])
        x = x + step_size * q
        t += 1
    return x, func([x,y])

# Create a list of values $x$
# Create a list of values $y$
step = 0.01
x0 = np.arange(-5, 5, step)
x1 = np.arange(-5, 5, step)

# Create a grid of $x$ and $y$ using 'np.meshgrid'
x0, x1 = np.meshgrid(x0, x1)

# Evaluate a function $f(x,y)$
z = sphere([x0, x1])

x_best, y_best = LS(sphere, sphere_derivative, S=[-0.5, 0.5], step_size=0.001, t_max=100)
print("Maximum value of f(x*) =", y_best)
print("Optimal solution x* =", x_best)

# Plot the function using 'plt.plot_surface'
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(x0, x1, z, cmap = cm.viridis, antialiased = True)

# Formatting the plot
ax.set_xlabel("x[0]")
ax.set_ylabel("x[1]")
ax.set_title("Sphere function")
```
